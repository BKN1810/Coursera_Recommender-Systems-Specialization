```python
import pandas as pd
import numpy as np
# sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
user_ratings = pd.read_excel('./Metric-SpreadSheet-Assignment.xlsx',skip_footer=70,skiprows=1,index_col=0)
```

```python
user_ratings = user_ratings.iloc[:,:-3]
```

```python
user_predictings = pd.read_excel('./Metric-SpreadSheet-Assignment.xlsx',skip_footer=56,skiprows=15,index_col=0)
```

```python
user_predictings = user_predictings.iloc[:,:-3]
```

```python
user_predictings.columns = user_ratings.columns
```

```python
MAE_by_user = pd.read_excel('./Metric-SpreadSheet-Assignment.xlsx',skiprows=27,skip_footer=44,index_col=0)
```

```python
MAE_by_user = MAE_by_user.iloc[:,:-3]
```

```python
MAE_by_user.columns = user_ratings.columns
```

# 1.Compute the mean absolute error for each user

```python
MAE_by_user['MAE'] = MAE_by_user.mean(axis=1)
```

```python
MAE_by_user.sort_values('MAE',ascending=False)
```

# 2. Also compute the mean absolute error associated with the predictions for each movie. For example, Forrest Gump has an MAE of 0.5.

```python
MAE_by_user.mean(axis=0).sort_values(ascending=False)
```

# 3. Compute the overall MAE three ways (three decimal places is enough):

```python
movie_names = ['11: Star Wars: Episode IV - A New Hope (1977)',
       '12: Finding Nemo (2003)', '13: Forrest Gump (1994)',
       '14: American Beauty (1999)',
       '22: Pirates of the Caribbean: The Curse of the Black Pearl (2003)',
       '24: Kill Bill: Vol. 1 (2003)',
       '38: Eternal Sunshine of the Spotless Mind (2004)',
       '63: Twelve Monkeys (a.k.a. 12 Monkeys) (1995)', '77: Memento (2000)',
       '85: Raiders of the Lost Ark (Indiana Jones and the Raiders of the Lost Ark) (1981)',
]
```

```python
print('normal MAE',MAE_by_user[movie_names].sum().sum()/MAE_by_user[movie_names].count().sum())
```

```python
print('by user MAE',MAE_by_user.MAE.mean())
```

```python
print('by movie MAE',MAE_by_user[movie_names].mean(axis=0).mean())
```

# 4. Next, create a matrix with the squared errors (instead of absolute errors), and compute the MSE and RMSE (root-mean-squared-error) for each user. 

```python
SE_by_user = (user_predictings[movie_names] - user_ratings[movie_names]).applymap(lambda x:np.square(x))
```

```python
SE_by_user['MSE'] = SE_by_user.mean(axis=1)
SE_by_user['RMSE'] = SE_by_user['MSE'].apply(lambda x:np.sqrt(x))
```

```python
SE_by_user.sort_values('MSE',ascending=False)
```

```python
SE_by_user.sort_values('RMSE',ascending=False)
```

# 5. Then, compute the overall MSE and RMSE across all predictions (just the normal way, no need to compute by-user or by-movie). Again, remember to average all the squared errors to compute MSE, then take the square root of the MSE to get RMSE.

```python
print('overall MSE',SE_by_user[movie_names].sum().sum()/SE_by_user[movie_names].count().sum())
```

```python
print('overall RMSE',np.sqrt(SE_by_user[movie_names].sum().sum()/SE_by_user[movie_names].count().sum()))
```
